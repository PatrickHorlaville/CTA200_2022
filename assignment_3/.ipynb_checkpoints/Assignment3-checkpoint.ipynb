## Question 1

```python
# Setting parameters & c grid

N = 100 # resolution of the c grid is NxN
iters = 100 # number of desired iterations on each z
z0 = 0 # initial value for the iteration computation

x = np.linspace(-2, 2, N)
y = np.linspace(-2, 2, N)

c = [[0 for j in range(len(y))] for i in range(len(x))]

for i in range(len(x)):
    for j in range(len(y)):
        
        c[i][j] = complex(x[i], y[j])
```

```python
vals = [[0 for j in range(len(y))] for i in range(len(x))]

for i in range(len(x)):
    for j in range(len(y)):
        
        vals[i][j] = iterate(z0, iters, c[i][j])
```

```python
res_iter = np.transpose(np.transpose(vals)[0])
num_steps = np.transpose(np.transpose(vals)[1])
bools = booling(x, y, res_iter)
```

```python
X, Y = np.meshgrid(x, y)

plt.contourf(X, Y, bools, cmap = 'binary')
plt.title("Converging [black] and diverging [grey] iterations")
plt.xlabel("X")
plt.ylabel("Y")
```

```python
X, Y = np.meshgrid(x, y)

plt.contourf(X, Y, num_steps, levels = 50)
plt.title(r"$c = x + iy$ grid")
plt.xlabel("X")
plt.ylabel("Y")
plt.colorbar(label = "Number of iterations before diverging")
plt.ylim([0, 2])
```

## Question 2

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors
from scipy.integrate import odeint
```

1. Setting up the equations

```python
def eqns(var, t, params):
    
    '''
    Sets up the differential equations of our system of ODEs.
    
    Parameters
    ----------
    
    var : array_like
          list of variables to our ODEs
          
    t : array_like
        array of the time steps over which we will integrate our ODEs
        
    params : array_like
             list of parameters involved in our ODEs
             
    Returns 
    -------
    
    f : array_like
        list of our different ODEs
    
    '''
    
    x, y, z = var
    sigma, r, b = params
    
    f = [-sigma*(x - y), r*x - y - x*z, -b*z + x*y]

    return f
```

```python
W0 = [0, 1, 0] # X0, Y0, Z0
srb = [10, 28, 8/3] # sigma, r, b
t = np.linspace(0, 60, 6000) # time interval
```

2. Solving the system of ODEs (here using odeint())

```python
sol = odeint(eqns, W0, t, args=(srb,))
```

3. Reproducing Figure 1 from Lorenz

```python
solns = np.transpose(sol)
x_evol = solns[0]
y_evol = solns[1]
z_evol = solns[2]
```

```python
plt.plot(t[0:1000], y_evol[0:1000])
plt.title('Numerical solution of the ODE from t=0 to t=10 (1000 steps) for y(t)')
plt.xlabel('t')
plt.ylabel('y(t)')
```

```python
plt.plot(t[1000:2000], y_evol[1000:2000])
plt.title('Numerical solution of the ODE from t=10 to t=20 (1000 steps) for y(t)')
plt.xlabel('t')
plt.ylabel('y(t)')
```

```python
plt.plot(t[2000:3000], y_evol[2000:3000])
plt.title('Numerical solution of the ODE from t=20 to t=30 (1000 steps) for y(t)')
plt.xlabel('t')
plt.ylabel('y(t)')
```

5. Repeating with slightly different set of initial conditions

```python
W0_p = [0, 1.00000001, 0] # X0', Y0', Z0'
srb = [10, 28, 8/3] # sigma, r, b
t = np.linspace(0, 60, 6000) # time interval
```

```python
sol_p = odeint(eqns, W0_p, t, args=(srb,))
diff = sol_p - sol
dist = np.sqrt(np.sum(diff**2, axis = 1))
```

```python
plt.plot(t[0:3000], dist[0:3000])
plt.yscale("log")
plt.title(r"Distance between $W$ and $W'$ as function of time")
plt.xlabel("t")
plt.ylabel("Distance between $W$ and $W'$")
```
